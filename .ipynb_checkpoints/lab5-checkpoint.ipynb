## Covariance 

```python
x.cov()
```

```python
x_cov=x.cov()
x_cov
```

```python
values,vectors=np.linalg.eigh(x_cov)
values,vectors
```

## Sorting the Eigen values

```python
index=values.argsort()[::-1]
index
```

sorting according to the index order

```python
values=values[index]
vectors=vectors[:,index]
values,vectors
```
